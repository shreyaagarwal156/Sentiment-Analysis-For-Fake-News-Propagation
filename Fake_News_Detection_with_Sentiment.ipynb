```python
def wordopt(text):
    text=text.lower()
    text= re.sub('\[.*?\]','',text)
    text=re.sub("\\W"," ",text)
    text= re.sub('https?://\S+|www\.\S+','',text)
    text=re.sub('<.*?>+','',text)
    text=re.sub('[%s]'% re.escape(string.punctuation),'',text)
    text=re.sub('\n','',text)
    text=re.sub('\w*\d\w*','',text)
    return text
        
```

```python
data['text']= data['text'].apply(wordopt)
```

```python
x= data['text']
y=data['class']
```

```python
from sklearn.linear_model import LogisticRegression
LR= LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr= LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
from Crypto.Random import get_random_bytes
import hashlib
from textblob import TextBlob
import base64
```

```python
key = get_random_bytes(16) 
```

```python
def encrypt_text(text):
    cipher = AES.new(key, AES.MODE_CBC)
    ct_bytes = cipher.encrypt(pad(text.encode(), AES.block_size))
    iv = cipher.iv
    ct = ct_bytes
    encrypted = base64.b64encode(iv + ct).decode('utf-8')
    return encrypted
```

```python
def decrypt_text(encrypted):
    raw = base64.b64decode(encrypted)
    iv = raw[:16]
    ct = raw[16:]
    cipher = AES.new(key, AES.MODE_CBC, iv)
    pt = unpad(cipher.decrypt(ct), AES.block_size)
    return pt.decode('utf-8')
```

```python
def hash_text(text):
    return hashlib.sha256(text.encode()).hexdigest()
```

```python

# 🧪 Apply AES and SHA-256 to first article
sample_text = data_merge['text'].iloc[0]

encrypted = encrypt_text(sample_text)
decrypted = decrypt_text(encrypted)
hashed = hash_text(sample_text)
print("Original Text Snippet:", sample_text[:100])
print("Encrypted:", encrypted[:100], "...")
print("Decrypted Snippet:", decrypted[:100])
print("SHA-256 Hash:", hashed)
```

```python

from textblob import TextBlob

def get_sentiment(news):
    blob = TextBlob(news)
    polarity = blob.sentiment.polarity
    if polarity > 0.1:
        return "Positive"
    elif polarity < -0.1:
        return "Negative"
    else:
        return "Neutral"
sentiment = get_sentiment(news)
print(f"Sentiment: {sentiment}")


```
